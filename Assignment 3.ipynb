```python
#nth term of fibonacci series
def fibonacci(x, y):
    a = 0
    b = 1
    l = []
    l.append(a)
    l.append(b)
    for i in range(0, x-3):
        c = a + b
        a = b
        b = c
        l.append(c)
    print('Fibonacci Series is', l)
    n = l[y - 1]
    print('The nth term in the fibonacci series is', n)
x = int(input('Enter the range of fibonacci series '))
y = int(input('Enter the nth position number in fibonacci series '))
fibonacci(x, y)
```

```python
#gcd by euclid's algorithm
def gcd(n1, n2):
    while n1 != 0 and n2 != 0:
        x = n1 % n2
        n1 = n2
        n2 = x
    if n1 == 0:
        print('GCD is', n2)
    else:
        print('GCD is', n1)

n1 = int(input('Enter the first number '))
n2 = int(input('Enter the second number '))
if n1 > n2:
    n1 = n1
    n2 = n2
    gcd(n1, n2)
elif n1 < n2 :
    a = n1
    n1 = n2
    n2 = a
    gcd(n1, n2)
else :
    print('Both numbers are equal')
```
